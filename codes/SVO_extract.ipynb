```python
import numpy as np
import pandas as pd
import textacy
import spacy
import re
import os
```

```python
path = '/Users/yuchenluo/Desktop/forecast_downturn/code_scripts/txt_sample/'
cols = ["docId", "subject","verb", "object"]
acceptable_labels = ["ORG", 'PER']
svos_org_per = []
NER_subjects = []

# use only ORG and PER ner
for filename in os.listdir(path):
    if filename.endswith('txt'):
        full_name = path + filename
        with open(full_name, 'r') as file:
            data = file.read()
            doc = textacy.make_spacy_doc(data, lang='en_core_web_sm')
            for ent in doc.ents:
                if ent.label_ in acceptable_labels:
                 # preprocess so NER tags match the subject text 
                    ner = ent.text.lower() #change to lower case
                    ner = re.sub(r'[^\w\s]', ' ', ner) # remove punct 
                    NER_subjects.append(ner)
            SVOs = textacy.extract.subject_verb_object_triples(doc)
            for s in SVOs:
                subject_list  = [i.text for i in s[0]] #convert token to list
                subject = " ".join(subject_list).lower() # list to string
                if subject in NER_subjects:
                    row = {"docId": filename, "subject": s[0], "verb":s[1], "object":s[2]}
                    svos_org_per.append(row)
                
svo_org_per = pd.DataFrame(svos_org_per,columns=cols)

```
