```python
import pandas as pd
import os
import json
```

```python
cols = ["docId", "text","NER"]
rows = []
acceptable_NER = ["ORGANIZATION", "PERSON"]
path = '/Users/yuchenluo/Downloads/stanford-postagger-full-2020-11-17/sample_output/'
  
# Parsing the all JSON files
for filename in os.listdir(path):
    if filename.endswith('json'):
        full_name = path + filename
        with open(full_name, 'r') as j:
            contents = json.loads(j.read())
            docId = contents['docId']
            sentences = contents['sentences']
            for sentence in sentences:
                mentions = sentence['entitymentions']
                for mention in mentions:
                    text = mention['text']
                    ner = mention['ner']
                    if ner in acceptable_NER:
                        rows.append({"docId": docId, "text": text, "NER": ner,})
df = pd.DataFrame(rows, columns=cols)
  
```

```python
cols = ["docId", "subject", "relation", "object"]
rows = []
path = '/Users/yuchenluo/Downloads/stanford-postagger-full-2020-11-17/sample_output/'
  
# Parsing the all JSON files
for filename in os.listdir(path):
    if filename.endswith('json'):
        full_name = path + filename
        with open(full_name, 'r') as j:
            contents = json.loads(j.read())
            docId = contents['docId']
            sentences = contents['sentences']
            for sentence in sentences:
                ies = sentence['openie']
                for ie in ies:
                    subject = ie['subject']
                    relation = ie['relation']
                    obj = ie["object"]
                    rows.append({"docId": docId,"subject": subject, "relation": relation, "object":obj})
df = pd.DataFrame(rows, columns=cols)
  
```

```python
# Writing dataframe to csv
df.to_csv('parsed_NER_grouped.csv')
```
